```python
from time import time

def find_max(data):
  biggest=data[0]
  for val in data:
    if val>biggest:
      biggest=val
  return biggest

n=int(input())
data=[]
for i in range(n):
  data.append(i)

start_time=time()
find_max(data)
end_time=time()
elapsed=end_time-start_time
print(elapsed)
```

```python
def prefix_average1(S):
  n=len(S)
  A=[0]*n
  for j in range(n):
    total=0
    for i in range(j+1):
      total+=S[i]
    A[j]=total/(j+1)
  return A

n=int(input())
S1=[]
for i in range(n):
  S1.append(i)

start_time=time()
prefix_average1(S1)
end_time=time()
elapsed=end_time-start_time
print(elapsed)
```

```python
def prefix_average2(S):
  n=len(S)
  A=[0]*n
  for j in range(n):
    A[j]=sum(S[0:j+1])/(j+1)
  return A

n=int(input())
S2=[]
for i in range(n):
  S2.append(i)

start_time=time()
prefix_average2(S2)
end_time=time()
elapsed=end_time-start_time
print(elapsed)
```
