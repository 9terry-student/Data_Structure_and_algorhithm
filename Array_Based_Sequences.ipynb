## 2. List

```python
# List.append

from time import time
def compute_average(n):
  data=[]
  start=time()
  for k in range(n):
    data.append(None)
  end=time()
  return (end-start)/n
i=100
while i<=100000000:
  print(compute_average(i))
  i*=10
```
